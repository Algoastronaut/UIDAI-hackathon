```python

# mapping states based on the pincode found
merged_df['state'] = merged_df['state'].replace('Darbhanga', 'Bihar')
merged_df['state'] = merged_df['state'].replace('Puttenahalli', 'Karnataka')
merged_df['state'] = merged_df['state'].replace('BALANAGAR', 'Telangana')
merged_df['state'] = merged_df['state'].replace('Uttaranchal', 'Uttarakhand')
merged_df['state'] = merged_df['state'].replace('Jaipur', 'Rajasthan')
merged_df['state'] = merged_df['state'].replace('Madanapalle', 'Andhra Pradesh')
merged_df['state'] = merged_df['state'].replace('Nagpur', 'Maharashtra')
merged_df['state'] = merged_df['state'].replace('Raja Annamalai Puram', 'Tamil Nadu')
```

```python
import re

def clean_state_name(state):
    """Clean and standardize state names."""
    if pd.isna(state):
        return state
    state = str(state).strip()            # Remove leading/trailing spaces
    state = state.lower()                 # Convert to lowercase
    state = re.sub(r'[^a-z\s]', '', state)  # Remove all non-letter characters
    state = ' '.join(state.split())       # Replace multiple spaces with single space
    return state
```

```python
# Standardized mapping for Indian state and union territory names
# This helps unify variations, typos, and alternative spellings into official names

state_standardization_map = {
    # ------------------------------
    # Northern States & UTs
    # ------------------------------
    'delhi': 'Delhi',
    'ladakh': 'Ladakh',
    'jammu and kashmir': 'Jammu and Kashmir',
    'jammu kashmir': 'Jammu and Kashmir',
    'himachal pradesh': 'Himachal Pradesh',
    'punjab': 'Punjab',
    'uttarakhand': 'Uttarakhand',
    'haryana': 'Haryana',
    'chandigarh': 'Chandigarh',
    'uttar pradesh': 'Uttar Pradesh',

    # ------------------------------
    # Western States
    # ------------------------------
    'rajasthan': 'Rajasthan',
    'goa': 'Goa',
    'gujarat': 'Gujarat',
    'maharashtra': 'Maharashtra',
    'nagpur': 'Maharashtra',  # Common city mapped to state

    # ------------------------------
    # Central States
    # ------------------------------
    'madhya pradesh': 'Madhya Pradesh',
    'chhattisgarh': 'Chhattisgarh',
    'chhatisgarh': 'Chhattisgarh',  # common typo

    # ------------------------------
    # Eastern States
    # ------------------------------
    'bihar': 'Bihar',
    'jharkhand': 'Jharkhand',
    'odisha': 'Odisha',
    'orissa': 'Odisha',
    'west bengal': 'West Bengal',
    'west bangal': 'West Bengal',
    'westbengal': 'West Bengal',
    'west  bengal': 'West Bengal',
    'westbengal ': 'West Bengal',
    'west bengli': 'West Bengal',
    'sikkim': 'Sikkim',
    'assam': 'Assam',
    'meghalaya': 'Meghalaya',
    'mizoram': 'Mizoram',
    'manipur': 'Manipur',
    'nagaland': 'Nagaland',
    'tripura': 'Tripura',
    'arunachal pradesh': 'Arunachal Pradesh',

    # ------------------------------
    # Southern States
    # ------------------------------
    'kerala': 'Kerala',
    'tamil nadu': 'Tamil Nadu',
    'andhra pradesh': 'Andhra Pradesh',
    'telangana': 'Telangana',
    'karnataka': 'Karnataka',

    # ------------------------------
    # Union Territories (Island / Coastal)
    # ------------------------------
    'andaman and nicobar islands': 'Andaman and Nicobar Islands',
    'andaman nicobar islands': 'Andaman and Nicobar Islands',
    'lakshadweep': 'Lakshadweep',
    'puducherry': 'Puducherry',
    'pondicherry': 'Puducherry',

    # ------------------------------
    # Special merged UTs
    # ------------------------------
    'dadra nagar haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'dadra and nagar haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'daman and diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'dadra and nagar haveli and daman and diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'the dadra and nagar haveli and daman and diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'daman diu': 'Dadra and Nagar Haveli and Daman and Diu',
}
```

```python

merged_df['state_clean'] = (
    merged_df['state']
    .apply(clean_state_name)
    .map(state_standardization_map)
    .fillna(merged_df['state'].apply(clean_state_name))
)
```

```python

# Drop invalid entries like numeric junk
merged_df = merged_df[~merged_df['state'].astype(str).str.isnumeric()]

# Optional: check unmapped states
unmapped = merged_df[merged_df['state_clean'].isna()]['state'].unique()
print("Unmapped values:", unmapped)
```
